3. What are the mean and standard deviation of flow_duration for each Attack_type, and are differences statistically significant?

```python
data["flow_duration"].mean()
```

```python
summ = sum(data["flow_duration"])
numm = len(data["flow_duration"])
mean_number = summ / numm
deviation_number = sum((x - mean_number) ** 2 for x in data["flow_duration"]) / (numm - 1)
deviation_data = deviation_number ** (1/2)
deviation_data
```
