```python
df = pd.read_csv('heart_2022_with_nans.csv')
df = pd.DataFrame(df)
df
```

```python
# Question: What is the correlation of health and physical activity?

health = df.groupby(['PhysicalActivities', 'HadHeartAttack'])['HadHeartAttack'].count()
health = health.plot(kind = 'bar', figsize = (10, 6), rot = 45)
plt.title('The Health Through Physical Activity')
plt.xlabel('Physical Activity')
plt.ylabel('Count')
plt.legend(title = '(Physical Activities, Had A Heart Attack)')
plt.show()

# Observation: The chart answered the question. Based on the data, it showed that those with no heart attack
# are physically active (majority of the respondents).
```

```python
# Question: What gender does have the highest number of heart attack incident?

heart = df.groupby(['Sex', 'HadHeartAttack'])['HadHeartAttack'].count()
heart = heart.plot(kind = 'bar', figsize = (10, 6), rot = 0)
plt.title('Heart Attack Incidence by Sex')
plt.xlabel('Heart Attack Status')
plt.ylabel('Count')
plt.legend(title = '(Sex, Had a Heart Attack)')
plt.show()

# Observation: The chart answered the question. Based on the data, it showed that male are much prone to the heart attack,
# and majority of the participants which are female have not had a heart attack.
```
