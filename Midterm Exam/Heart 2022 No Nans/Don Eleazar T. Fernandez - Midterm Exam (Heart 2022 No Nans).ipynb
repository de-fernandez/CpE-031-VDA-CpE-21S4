```python
df = pd.read_csv('heart_2022_no_nans.csv')
df = pd.DataFrame(df)
df
```

```python
# Question: Does vaccination lead to a better health for an individual?

vaccine = df.groupby(['GeneralHealth', 'FluVaxLast12'])['FluVaxLast12'].count().unstack()
vaccine = vaccine.plot(kind = 'bar', figsize = (10, 6), rot = 0)
plt.title('Flu Vaccination Status by General Health')
plt.xlabel('General Health')
plt.ylabel('Vaccine')
plt.legend(title = 'Flu Vaccination', labels = ['No', 'Yes'])
plt.tight_layout()
plt.show()

# Observation: The chart answered the question. Based on the data, it showed that those that most which are in "Good" and "Very Good" health statuses are more frequently vaccinated,
# and while the rest has a few time of vaccination. It can also be interpreted from the graph that vaccinated individuals tend to have a good health status.
```

```python
# Question: Can the drink of an alcohol trigger a heart stroke?

heart = df[df['HadHeartAttack'] == 'Yes']
heart = heart.groupby('AlcoholDrinkers')['HadHeartAttack'].count()
plt.figure(figsize = (10, 6))
plt.pie(heart, labels = heart.index, autopct = '%1.1f%%', startangle = 90)
plt.title('The Heart Attack Among People That Drink Alcohol.')
plt.show()

# Observation: The chart answered the question. Based on the data,
# it showed that the drink of alcohol does not lead or trigger a heart stroke. In the data, 60.4% answered "No" and 39.6% answered "Yes".
```
